## Seleções e Frequências

```python
df = pd.read_csv('aluguel_residencial.csv', sep = ';')
df
```

```python
# 1. Selecione somente os imóveis classificados com o tipo 'Apartamento':
selecao = df['Tipo'] == 'Apartamento'
n1 = df[selecao].shape[0]
n1
```

```python
# 2. Selecione os imóveis classificados com tipo 'Casa', 'Casa de Condominío' e 'Casa de Vila':
selecao = (df['Tipo'] == 'Casa') | (df['Tipo'] == 'Casa de Condominío') | (df['Tipo'] == 'Casa de Vila')
n2 = df[selecao].shape[0]
n2
```

```python
# 3. Selecione os imóveis entre 60 e 100 metros quadrados, incluindo os limites:
selecao = (df['Area'] >= 60) & (df['Area'] <= 100)
n3 = df[selecao].shape[0]
n3
```

```python
# 4. Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00:
selecao = (df['Quartos'] >= 4) & (df['Valor'] < 2000)
n4 = df[selecao].shape[0]
n4
```

```python
print("Número de imóveis classificados com o tipo 'Apartamento'-> {}".format(n1))
print("Número de imóveis classificados com tipo 'Casa', 'Casa de Condominío' e 'Casa de Vila'-> {}".format(n2))
print("Número de imóveis entre 60 e 100 metros quadrados, incluindo os limites -> {}".format(n3))
print("Número de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00 -> {}".format(n4))
```

# Relatório de Análise VIII

## Identificando e Removendo Outlers

```python
df = pd.read_csv('aluguel_residencial.csv', sep = ';')
```

```python
df.boxplot(['Valor'])
```

```python
df.query('Valor >= 500000')
```

```python
valor = df['Valor']
```

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
selecao = (valor >=limite_inferior) & (valor <= limite_superior)
df_new = df[selecao]
```

```python
df_new.boxplot(['Valor'])
```

```python
df.hist(['Valor'])
df_new.hist(['Valor'])
```

```python
df.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = df.groupby('Tipo')['Valor']
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
Q1
```

```python
limite_superior['Apartamento']
```

```python
df_new = pd.DataFrame()

for tipo in grupo_tipo.groups.keys():
    eh_tipo = df['Tipo'] == tipo
    eh_dentro_limite = (df['Valor'] >= limite_inferior[tipo]) & (df['Valor'] <= limite_superior[tipo])
    selecao = eh_tipo & eh_dentro_limite
    df_selecao = df[selecao]
    df_new = pd.concat([df_new, df_selecao])
    
```

```python
df_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
df_new.to_csv('aluguel_residencial_sem_outliers.csv', sep = ';', index = False)
```
